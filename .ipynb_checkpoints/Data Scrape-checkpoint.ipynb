```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# NBA season we will be analyzing
# URL page we will scraping (see image above)
#url = "https://www.basketball-reference.com/leagues/NBA_{}_per_game.html".format(year)
url_2019 = "https://www.basketball-reference.com/players/h/hardeja01/splits/2019"
# this is the HTML from the given URL
html = urlopen(url_2019)
soup = BeautifulSoup(html)


tables = pd.read_html(url_2019)

Harden2019_1 = tables[0]


#Clean up the header to remove the "unnamed, totals, advanced, etc; tags"
pd.set_option('display.max_rows',Harden2019_1.shape[0]+1)

Harden2019_1.columns = Harden2019_1.iloc[0]
Harden2019_2 = Harden2019_1[1:]
Harden2019_2.columns = Harden2019_2.iloc[0]
Harden2019_3 = Harden2019_2[1:]

#Break down the data by the splits in the table. 
#Stats vs. individual opponents

Harden2019_byteam = Harden2019_3.iloc[46:,:]
Harden2019_byteam = Harden2019_byteam.reset_index()
Harden2019_byteam = Harden2019_byteam.iloc[:,2:]

#Add column for year
Harden2019_byteam['Year'] = 2019

Harden2019_byteam

Harden2019_byteam.to_csv('Harden_by_team_2019.csv',index=False)


#If you're wondering why I do what I do below, it's because the scraped DF has duplicate column names but they hold 
#differing data I feel it's WAY faster to CSV the file then reload it ASAP as a new DF. 
#Pandas will automatically account for the dups

Harden2019 = "Harden_by_team_2019.csv"
Harden2019_byteam = pd.read_csv(Harden2019, encoding="ISO-8859-1")
Harden2019_byteam = Harden2019_byteam.rename(columns={"MP.1":"AvMP","PTS.1":"AvPTS","TRB.1":"AvRB","AST.1":"AvAST"})

Harden2019_byteam
```
